```python
book_df=book_df.drop("Unnamed: 0",axis=1)
```

```python
book_df=book_df.rename({'User.ID' :'User_ID',
                           'Book.Title': 'Book_title',
                           'Book.Rating':'Book_rating'},axis=1)
```

```python
user_book_df=book_df.pivot(index='User_ID', columns='Book_title', values='Book_rating')
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

# CORRELATION similarity

```python
user_sim_corr = 1 - pairwise_distances( user_book_df.values,metric='correlation')
user_sim_corr
```

```python
user_sim_corr_df.sort_values([9],ascending=False)
```

```python
def give_reco_correlation(customer_id):
    tem=list(user_sim_df.sort_values([customer_id],ascending=False).head(100).index)
    book_list=[]
    for i in tem:
        book_list=book_list+list(book_df[book_df["User_ID"]==i]["Book_title"])
    return set(book_list)-set(book_df[book_df["User_ID"]==customer_id]["Book_title"])
```

```python
# Recommendation using correlation similarity
give_reco_correlation(9)
```
